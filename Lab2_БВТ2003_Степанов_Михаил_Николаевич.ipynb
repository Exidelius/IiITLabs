```python
def binarySearch(arr: list[int], value: int, left: int, right: int):
    
    if right>=left:
        middle = (left+right)//2
    
        if (arr[middle]==value):
            return middle
        elif (arr[middle]>value):
            return binarySearch(arr, value, left, middle-1)
        elif (arr[middle]<value):
            return binarySearch(arr, value, middle+1, right)
    else:
        return None
```

```python
class BinarySearchTreeNode:
    def __init__(self, value: float | int = None, index: int | set[int] = None):
        self.left = None
        self.right = None
        
        self.value = value
        self.index: set[int]
        
        if type(index) == set:
            self.index = index
        elif type(index) == int:
            self.index = {index}

    def insert(self, value):
        if self == value:
            self.index.update(value.index)
            return
        if value < self:
            if self.left != None:
                self.left.insert(value)
                return
            self.left = value
        else:
            if self.right != None:
                self.right.insert(value)
                return
            self.right = value

    def __eq__(self, other):
        if type(other) == BinarySearchTreeNode:
            return self.value == other.value
        return False

    def __lt__(self, other):
        if type(other) == BinarySearchTreeNode:
            return self.value < other.value
        return False

    def __gt__(self, other):
        if type(other) == BinarySearchTreeNode:
            return self.value > other.value
        return False

    def search(self, other):
        if self == other:
            return self
        if self is not None and other < self:
            return self.left.search(other)
        if self is not None and other > self:
            return self.right.search(other)
        return None
```

```python
class BinarySearchTree:
    def __init__(self, array: list[int | float] = None):
        self.array = sorted(array) if array is not None else []
        self.middle = len(self.array) // 2
        self.tree = BinarySearchTreeNode(self.array[self.middle], self.middle)
        for index, element in enumerate(self.array):
            self.tree.insert(BinarySearchTreeNode(element, index))

    def search(self, value: int | float):
        search_result = self.tree.search(BinarySearchTreeNode(value, 0))
        if search_result is not None:
            return search_result.index
        return None
```

```python
def fibonacciSearch(arr, value):
    f0 = 0
    f1 = 1
    f2 = 1
    while(f2<len(arr)):
        f0=f1
        f1=f2
        f2=f1+f0

    start_index = -1
    while(f2>1):
        index = min(start_index+f0, len(arr)-1)
        if arr[index] < value:
            f2=f1
            f1=f0
            f0=f2-f1
            start_index = index
        elif arr[index] > value:
            f2=f0
            f1-=f0
            f0=f2-f1
        else: 
            return index
    if (arr[len(arr)-1]==value):
        return len(arr)-1
    return None
```

```python
def interpolationSearch(arr, value, left, right):
    if (left<=right and value >= arr[left] and value <= arr[right]):
        index = left + ((right - left)//(arr[right]-arr[left])*(value-arr[left]))

        if arr[index] == value:
            return index

        if arr[index]<value:
            return interpolationSearch(arr, value, index-1, right)

        if arr[index]>value:
            return interpolationSearch(arr, value, left, index+1)

    return None
```

Простое рехэширование

```python
class HashTable:
    def __init__(self, capacity=10):
        self.capacity = capacity
        self.massive = [None for i in range(self.capacity)]
        self.length = 0

    def rehash(self, hashed_value):
        return (hashed_value + 1) % self.capacity

    def add(self, key, value): 
        hashed_key = hash(key)
        index = hashed_key % self.capacity

        if self.capacity != self.length:
            while (self.massive[index] != None):
                index = self.rehash(index)
            self.massive[index] = (key, value)
            self.length+=1
        else:
            print('Недостаточно места')
        

    def get(self, input_key):
        hashed_key = hash(input_key)
        index = hashed_key % self.capacity
        counter = 0

        while self.massive[index] != None and counter != self.capacity:
            if self.massive[index][0] == input_key:
                return self.massive[index][1]
            else:
                index = self.rehash(index)
                counter+=1
        
        return None

    def __str__(self):
        return str(self.massive)
```

Метод цепочек

```python
class HashTableChain:
    def __init__(self, capacity=10):
        self.capacity = capacity
        self.buckets = [[] for i in range(self.capacity)]
        self.length = 0

    def put(self, key, value): 
        hashed_key = hash(key)
        index = hashed_key % self.capacity

        self.buckets[index].append((key,value))


    def get(self, input_key):
        hashed_key = hash(input_key)
        index = hashed_key % self.capacity
        bucket = self.buckets[index]

        for key, value in bucket:
            if key == input_key:
                return(value)

        return None

    def __str__(self):
        return str(self.buckets)
```

## Задание №3
Расставить на стандартной 64-клеточной шахматной доске 8 ферзей так, чтобы ни один из них не находился под боем другого». Подразумевается, что ферзь бьёт все клетки, расположенные по вертикалям, горизонталям и обеим диагоналям
Написать программу, которая находит хотя бы один способ решения задач.


```python
all_x_coords = []
all_y_coords = []
```

```python
def isContains(array: list[int], value: int):
    for element in array:
        if element == value:
            return True
    return False

def isCoordsContains(all_x: list[list[int]], all_y: list[list[int]], x_coords: list[int], y_coords: list[int]):
    for i in range(len(all_x)):
        is_equal = True
        for j in range(8):
            if all_y[i][all_x[i].index(j)] != y_coords[x_coords.index(j)]:
                is_equal=False
                break

        if is_equal:
            return True
    return False
```

```python
def isPositionOnDiagonal(x_coords, y_coords, current_x, current_y):
    for i in range(len(x_coords)):
        if (abs(x_coords[i] - current_x) - abs(y_coords[i] - current_y) == 0):
            return True

    return False    
```

```python
def findQueenPosition(x_coords: list[int], y_coords: list[int], counter: int):
    if counter<8:
        
        for i in range(8):
            if not isContains(x_coords, i):
                
                for j in range(8):
                    if not isContains(y_coords, j):
                        
                        if not isPositionOnDiagonal(x_coords, y_coords, i,j):
                            x_coords_copy = x_coords.copy()
                            x_coords_copy.append(i)
                            y_coords_copy = y_coords.copy()
                            y_coords_copy.append(j)
                            findQueenPosition(x_coords_copy, y_coords_copy, counter+1)
    else: 
        if len(all_x_coords)>0:           
            if not isCoordsContains(all_x_coords, all_y_coords, x_coords, y_coords):         
                all_x_coords.append(x_coords)
                all_y_coords.append(y_coords)
        else:
            all_x_coords.append(x_coords)
            all_y_coords.append(y_coords)
```

```python
def queensPositionsCountOnCoords(x,y):
    x_coords=[]
    x_coords.append(x)
    y_coords=[]
    y_coords.append(y)
    
    return findQueenPosition(x_coords, y_coords, 1)
```
